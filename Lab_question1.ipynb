```python
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

sizes = [1000, 5000, 10000, 20000, 50000]

print("Linear Search Time Analysis")
print("Size\tBest Case\tAverage Case\tWorst Case")
print("-" * 50)

best_times = []
avg_times = []
worst_times = []

for n in sizes:
    arr = list(range(1, n+1))
    random.shuffle(arr)

    start = time.time()
    linear_search(arr, arr[0])
    best_time = (time.time() - start) * 1000
    best_times.append(best_time)

    start = time.time()
    linear_search(arr, arr[n//2])
    avg_time = (time.time() - start) * 1000
    avg_times.append(avg_time)

    start = time.time()
    linear_search(arr, n+1000)
    worst_time = (time.time() - start) * 1000
    worst_times.append(worst_time)

    print(f"{n}\t{best_time:.4f} ms\t{avg_time:.4f} ms\t{worst_time:.4f} ms")

plt.figure(figsize=(10, 6))
plt.plot(sizes, best_times, 'o-', label='Best Case O(1)', linewidth=2)
plt.plot(sizes, avg_times, 's-', label='Average Case O(n/2)', linewidth=2)
plt.plot(sizes, worst_times, '^-', label='Worst Case O(n)', linewidth=2)
plt.xlabel('Input Size (n)')
plt.ylabel('Time (milliseconds)')
plt.title('Linear Search Time Analysis')
plt.legend()
plt.grid(True)
plt.show()

print("\nTime Complexity Analysis:")
print("Best Case: O(1) - Element found at first position, constant time")
print("Average Case: O(n/2) - Element found in middle, linear growth")
print("Worst Case: O(n) - Element not found, must check all elements")

print(f"\nObservations:")
print(f"- Best case remains nearly constant: {min(best_times):.4f} to {max(best_times):.4f} ms")
print(f"- Worst case grows linearly: {min(worst_times):.4f} to {max(worst_times):.4f} ms")
print(f"- Time ratio (largest/smallest): {max(worst_times)/min(worst_times):.2f}x increase")
```
