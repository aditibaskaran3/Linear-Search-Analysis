```python
import random
import matplotlib.pyplot as plt

def linear_search_count(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons
    return -1, comparisons

sizes = [100, 500, 1000, 5000, 10000]

print("Linear Search Comparison Count Analysis")
print("Size\tBest Case\tAverage Case\tWorst Case")
print("-" * 50)

best_comps = []
avg_comps = []
worst_comps = []

for n in sizes:
    arr = list(range(1, n+1))
    random.shuffle(arr)

    _, best_comp = linear_search_count(arr, arr[0])
    best_comps.append(best_comp)

    _, avg_comp = linear_search_count(arr, arr[n//2])
    avg_comps.append(avg_comp)

    _, worst_comp = linear_search_count(arr, n+100)
    worst_comps.append(worst_comp)

    print(f"{n}\t{best_comp}\t\t{avg_comp}\t\t{worst_comp}")

plt.figure(figsize=(10, 6))
plt.plot(sizes, best_comps, 'o-', label='Best Case O(1)', linewidth=2)
plt.plot(sizes, avg_comps, 's-', label='Average Case O(n/2)', linewidth=2)
plt.plot(sizes, worst_comps, '^-', label='Worst Case O(n)', linewidth=2)
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Comparisons')
plt.title('Linear Search Comparison Count Analysis')
plt.legend()
plt.grid(True)
plt.show()

print("\nTheoretical vs Empirical Analysis:")
print("Size\tTheoretical\t\tEmpirical")
print("\tBest\tAvg\tWorst\tBest\tAvg\tWorst")
print("-" * 60)

for i, n in enumerate(sizes):
    print(f"{n}\t1\t{n//2}\t{n}\t{best_comps[i]}\t{avg_comps[i]}\t{worst_comps[i]}")

print("\nComparison Growth Analysis:")
print("Best Case: Always 1 comparison regardless of input size")
print("Average Case: Approximately n/2 comparisons, grows linearly")
print("Worst Case: Exactly n comparisons, perfect linear growth")

print(f"\nLinear Growth Verification:")
for i in range(1, len(sizes)):
    size_ratio = sizes[i] / sizes[i-1]
    worst_ratio = worst_comps[i] / worst_comps[i-1]
    print(f"Size {sizes[i-1]} to {sizes[i]}: {size_ratio:.1f}x size, {worst_ratio:.1f}x comparisons")
```
